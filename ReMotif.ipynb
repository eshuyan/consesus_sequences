```python
def Motif_Search(protein_sequence, motif):
    ''' Input: protein_sequence, string all uppercase
               motif, a regular express
        Oouput: number of appearance and locations
        '''
    found = re.findall(motif,protein_sequence)
    found.append(re.findall(motif,protein_sequence[::-1]))
    if len(found) == 0:
        return False
    else:
        return found
    
```

```python
def parse_fasta(F):
    '''Input: a fasta file
    Output: libary key == ID, value == sequence'''
   
    F = F.readlines()
    protein_name = []
    sequence_start = []
    i = 0
    while i < (len(F)-1):
        
        if F[i][0] == '>':
            protein_name.append(F[i].strip('\n'))
            sequence_start.append(i+1)
        i += 1
    sequence_start.append(len(F))
    protein = {}

    for j in range(len(protein_name)-1):
        seq = F[sequence_start[j]:sequence_start[j+1]]
        sequences = ''
        for line in seq:
            sequences += line.strip(' ').strip('\n')
        protein[protein_name[j]] = sequences
    
    return protein 
```

```python
F = open('sequence.fasta','r')
worm_protein = parse_fasta(F)
```
